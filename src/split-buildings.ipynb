# Split buildings

Break the building points into batches for processing

```python
import os
import geopandas as gpd
```

```python
%store -r
```

```python
points_path = os.path.join(output_dir, "california-building-points.shp")
```

```python
points = gpd.read_file(points_path)
```

```python
batch_size = 10000
```

```python
end = 0 + batch_size
for i, start in enumerate(range(0, points.shape[0], batch_size)):
    chunk_df = points[start:end]
    print(f"Writing batch {i+1} with rows {start}-{end}")
    chunk_df.to_file(f"{output_dir}/california-building-points-batch-{i+1}.shp")
    end += batch_size
```
